### Importação dos dados

#1. Análise do faturamento


Calculando o faturamento total de cada loja.


```python
faturamento_loja = loja["Preço"].sum()
faturamento_loja2 = loja2["Preço"].sum()
faturamento_loja3 = loja3["Preço"].sum()
faturamento_loja4 = loja4["Preço"].sum()
print(f"O faturamento da loja 1 foi de R${faturamento_loja}")
print(f"O faturamento da loja 2 foi de R${faturamento_loja2}")
print(f"O faturamento da loja 3 foi de R${faturamento_loja3}")
print(f"O faturamento da loja 4 foi de R${faturamento_loja4}")
```

Calculando o faturamento total de todas as lojas.

```python
lista_faturamentos = [faturamento_loja, faturamento_loja2, faturamento_loja3, faturamento_loja4]
nomes_lojas = ["Loja 1", "Loja 2", "Loja 3", "Loja 4"]
```

```python
faturamento_total = sum(lista_faturamentos)
print(f"O faturamento total foi de R${faturamento_total}")
```

Calculando a média dos faturamentos das lojas.

```python
media_faturamento = sum(lista_faturamentos) / len(lista_faturamentos)
print(f"A média de faturamento foi de R${media_faturamento}")
```

Calculando o maior e o menor faturamento.

```python
faturamento_max = max(lista_faturamentos)
faturamento_min = min(lista_faturamentos)
print(f"Maior faturamento: R$ {faturamento_max:.2f}")
print(f"Menor faturamento: R$ {faturamento_min:.2f}")
```

```python
indice_menor = lista_faturamentos.index(faturamento_min)
loja_menor = f"Loja {indice_menor + 1}"
print(f"A {loja_menor} tem o menor faturamento")
indice_maior = lista_faturamentos.index(faturamento_max)
loja_maior = f"Loja {indice_maior + 1}"
print(f"A {loja_maior} tem o maior faturamento")
```

Calculando o percentual de cada loja sobre o valor total do faturamento.

```python
def calcular_percentual(valor, total):
  return round((valor / total) * 100, 2)
for i, faturamento in enumerate(lista_faturamentos):
    percentual = calcular_percentual(faturamento, faturamento_total)
    print(f"A {nomes_lojas[i]} representa {percentual}% do faturamento total")
```

```python
percentuais = [calcular_percentual(fat, faturamento_total) for fat in lista_faturamentos]
rotulos = [f'{nome}: {perc}%' for nome, perc in zip(nomes_lojas, percentuais)]

indice_menor = lista_faturamentos.index(min(lista_faturamentos))

explode = [0.01] * len(lista_faturamentos)  # Pequena separação para todas
explode[indice_menor] = 0.1                 # Separação maior para a menor loja

plt.figure(figsize=(10, 8))

plt.pie(
    lista_faturamentos,          # Valores para o gráfico
    labels=rotulos,              # Rótulos com nome e percentual
    explode=explode,             # Separação das fatias
    autopct='',                  # Não mostrar percentuais duplicados (já estão nos rótulos)
    startangle=90,               # Ângulo inicial
    shadow=True,                 # Sombra para efeito 3D
    colors=['#ff9999','#66b3ff','#99ff99','#ffcc99']  # Cores personalizadas
)

plt.title('Distribuição do Faturamento por Loja', fontsize=16, pad=20)

plt.figtext(0.5, 0.01,
           f'A {nomes_lojas[indice_menor]} apresenta o menor faturamento: R$ {lista_faturamentos[indice_menor]:,.2f}',
           ha='center', fontsize=12, bbox={"facecolor":"#f0f0f0", "alpha":0.5, "pad":5})

plt.axis('equal')
```
